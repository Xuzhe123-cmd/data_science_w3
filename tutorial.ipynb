```python
import numpy as np
```

```python
array_2x2 = np.array([[1, 2],
                      [3, 4]])
print("1. 2x2 Array:")
print(array_2x2)
```

```python
random_3x4 = np.random.rand(3, 4)
print("2. Random 3x4 Array:")
print(random_3x4)
```

```python
ones_array = np.ones((2, 3))
zeros_array = np.zeros((2, 3))
print("3. Array of Ones:")
print(ones_array)
print("\nArray of Zeroes:")
print(zeros_array)
```

```python
random_6x3 = np.random.rand(6, 3)
print("4. Random 6x3 Array:")
print(random_6x3)
print("\nTransposed Array (3x6):")
print(random_6x3.T)
```

```python
n, p, m = 2, 3, 4
A = np.random.rand(n, p)
B = np.random.rand(p, m)
dot_product = np.dot(A, B)

print("5. Matrix A (n x p):")
print(A)
print("\nMatrix B (p x m):")
print(B)
print("\nDot Product (n x m):")
print(dot_product)
```
